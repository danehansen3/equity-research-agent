```python
universe_df = pd.read_excel('/Users/danehansen/Desktop/comp-sci/agents/equity-research-agent/data/stock_universe.xlsx')
tickers = universe_df['ticker'].tolist()
```

```python
filing_metadata = []
filing_metadata = []

for ticker in tickers:
    c = Company(ticker)
    filing = c.get_filings(form="10-Q").latest(1)

    ### DOWNLOAD MARDOWN ###
    filing_info = filing.to_dict()
    metadata = {k: v for k, v in filing_info.items() if k in ['accession_no', 'form', 'filing_date']}
    filing_metadata.append(metadata)

    filing_entity = filing.get_entity().to_dict()
    for k, v in filing_entity.items():
        if k in ['cik', 'name', 'display_name', 'tickers', 'exchanges', 'industry',' category']:
            metadata[k] = v
    filing_metadata.append(metadata)
    with open(f"../data/mds/{metadata['tickers']}.md", "w") as f:
        f.write(filing.markdown())
```

```python
from datetime import date

# Extracting the filing dates
filing_dates = [item['filing_date'] for item in filing_metadata]

# Finding the oldest and most recent dates
oldest_date = min(filing_dates)
most_recent_date = max(filing_dates)

print("Oldest date:", oldest_date)
print("Most recent date:", most_recent_date)
```
